```python
import pandas as pd

import pickle
test=pd.read_csv('../data/output/test_X.csv')
test=test.drop(['Unnamed: 0'], axis=1)
test
```

```python
submission=pd.read_csv('../input/test.csv')
submission
```

```python
# Merge and check
test_X = pd.merge(test, submission, on=['shop_id', 'item_id'], how='left')
print(len(test_X))
test_X.head()
```

```python
test_X=test_X.drop_duplicates(subset=['shop_id', 'item_id'])
test_X
```

```python
test_X=test_X.drop(['ID'], axis=1)
test_X
```

```python
test_X.to_csv('../data/output/test_final.csv', index=False)
```

```python
import pickle

# Load first model (index starts from 0)
MODEL_CB = pickle.load(open('../models/CB_MODEL.pickle', "rb"))
```

```python
# Make predictions on train set
model_cb_pred= MODEL_CB.predict(test_X)

print('Model CB predictions:', model_cb_pred)
```

```python
sample_submission=pd.read_csv('../input/sample_submission.csv')
sample_submission
```

```python
sample_submission['item_cnt_month']=model_cb_pred
sample_submission
```

```python
sample_submission['item_cnt_month'].min()
```
